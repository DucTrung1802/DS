```python
def is_na(value):
    return value is None or (isinstance(value, float) and pd.isna(value))


def is_stop(constraints: List[Tuple[str, str]]) -> str:
    if is_na(constraints):
        return (None, None)  # or return some default value

    for item in constraints:
        if item[0] in {"Repair", "Exam", "Overhaul"}:
            return (item[0], item[1])

    return (None, None)


def is_k1(constraints: List[Tuple[str, str]]) -> str:
    if is_na(constraints):
        return None  # or return some default value

    for item in constraints:
        if item[0] == "K1":
            return item[1]

    return None


def is_complete():
    return not schedule.isnull().values.any()
```

```python
# Main

dates.sort(reverse=True)


def recursion() -> bool:
    if is_complete():
        return True

    for train in trains:
        # Dates are in reverse order
        for day in dates:
            if not is_na(constraints.loc[train, day]):
                continue

            used_trips = []

            # Check stop place ("Repair", "Exam", "Overhaul")
            reason, stop_place = is_stop(constraints.loc[train, day])
            if stop_place:
                schedule.loc[train, day] = (reason, stop_place)
                if recursion():
                    return True
                else:
                    return False

            # Check K1
            k1_place = is_k1(constraints.loc[train, day])
            if k1_place:
                for trip in trips[day]:
                    if trip[2] == k1_place and trip not in used_trips:
                        schedule.loc[train, day] = trip
                        if recursion():
                            return True
                        else:
                            used_trips.append(trip)

                # Cannot find eligible trip for K1
                return False

            for trip in trips[day]:
                if trip not in used_trips:
                    schedule.loc[train, day] = trip
                    if recursion():
                        return True
                    else:
                        used_trips.append(trip)

            return False
```
