## 1.2. Tạo cột Customer gồm Young (18-30 tuổi), Senior (trên 30-50 tuổi), Elderly (trên 50 tuổi)


```python
# 1.2.
def categorize_age(age):
    if age >= 18 and age <= 30:
        return 'Young'
    elif age > 30 and age <= 50:
        return 'Senior'
    else:
        return 'Elderly'

df1_change_name_index['Customer'] = df1_change_name_index['Age'].apply(lambda x: categorize_age(x))
```

- Customer, 
- Class Name nhận được nhiều lượt review nhất trong nhóm Customer đó, 
- số lượng review tương ứng.


```python
# 1.3.
new_df1_1 = pd.DataFrame()
filter_ids = df1_change_name_index.groupby('Customer')['Positive_Feedback_Count'].idxmax()
new_df1_1["Customer"] = df1_change_name_index.loc[filter_ids]["Customer"]
new_df1_1["Class_Name"] = df1_change_name_index.loc[filter_ids]["Class_Name"]
new_df1_1["Reviews"] = df1_change_name_index.loc[filter_ids]["Positive_Feedback_Count"]
new_df1_1.head()
```
