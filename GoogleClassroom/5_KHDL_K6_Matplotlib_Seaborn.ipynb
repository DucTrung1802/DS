Bài 2: Phòng kinh doanh có 3 đội sale phụ trách kinh doanh 3 loại sản phẩm. Kết quả kinh doanh (tỷ VNĐ đồng) được tổng hợp như trong dữ liệu dưới đây.
Vẽ biểu đồ đa cột (multiple bar chart) để so sánh kết quả kinh doanh của 3 team cho các sản phẩm này

```python
import numpy as np
data = {
    "Team": ["Team 1", "Team 2", "Team 3"],
    "Mobile": [5.5, 10.3, 15.9],
    "Headphone": [15.4, 23.6, 30.2],
    "Laptop": [5.1, 9.2, 12.8],
}
df_2 = pd.DataFrame(data)


teams = data['Team']
mobile = data['Mobile']
headphone = data['Headphone']
laptop = data['Laptop']

x = np.arange(len(teams))  # the label locations
width = 0.25  # the width of the bars

fig, ax = plt.subplots()

ax.bar(x - width, mobile, width, label='Mobile')
ax.bar(x, headphone, width, label='Headphone')
ax.bar(x + width, laptop, width, label='Laptop')

ax.set_xlabel('Teams')
ax.set_ylabel('Usage')
ax.set_title('Usage of Devices by Team')
ax.set_xticks(x)
ax.set_xticklabels(teams)
ax.legend()

plt.show()
```

Bài 3: Với cùng bộ dữ liệu như Bài 2, vẽ biểu đồ tròn so sánh tổng doanh số (cho tất cả các sản phẩm) của 3 đội sale.

```python
import numpy as np

data = {
    "Team": ["Team 1", "Team 2", "Team 3"],
    "Mobile": [5.5, 10.3, 15.9],
    "Headphone": [15.4, 23.6, 30.2],
    "Laptop": [5.1, 9.2, 12.8],
}
df_2 = pd.DataFrame(data)


teams = data["Team"]
total = np.sum(data["Mobile"] + data["Headphone"] + data["Laptop"]).round(2)
mobile = sum(data["Mobile"]) / total
headphone = sum(data["Headphone"]) / total
laptop = sum(data["Laptop"]) / total

plt.title("Total revenues for each team")
plt.pie([mobile, headphone, laptop], labels=teams, autopct='%1.1f%%')
```

```python
# 4.5
labels = df.columns[1:-1]
total = np.sum(df["total_profit"]).round(2)
col_total_data = []
for label in labels:
    col_data = np.sum(df[label]) / total
    col_total_data.append(col_data)

plt.title("Venn diagram")
plt.pie(col_total_data, labels=labels, autopct="%1.1f%%")
```
